```r
library(dplyr)
mtcars %>%
    filter(mpg>20)
```

```r
mtcars %>%
select(mpg)
```

```r
mtcars %>%
mutate(
economy = ifelse(mpg>20, "Carro economico", "Carro gastador")
) %>%
select(mpg, economy)
```

```r
mtcars %>%
arrange(mpg) %>%
select(everything())
```

```r
mtcars %>%
summarise(media_mpg = median(mpg, na.rm=TRUE))
```
